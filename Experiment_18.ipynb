```
import numpy as np
import pandas as pd
np.random.seed(0)
df = pd.DataFrame({
    'Category': ['A','B','C','D','E'],
    'Values':[0,70,30,90,60],
    'Sales': np.random.randint(100,500,5),
    'Profit': np.random.randint(10,100,5)
})
print(df)
```

```
import matplotlib.pyplot as plt
#1. Area Plot
plt.figure()
plt.fill_between(df['Category'],df['Values'],color='green',alpha=0.25)
plt.title("Priyal")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid") # apply seaborn style
plt.figure(figsize=(7,4))
plt.fill_between(df['Category'],df['Values'],color='skyblue',alpha=0.5,label='Sales')
plt.fill_between(df['Category'],df['Profit'],color='orange',alpha=0.5,label='Profit')
plt.title("Priyal")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()
plt.show
```

```
#2. Pie Chart
plt.figure()
plt.pie(df['Values'],labels=df['Category'],autopct='%1.2f%%')
plt.title("Priyal")
plt.show()
```

```
#3. Donut Chart
plt.figure()
plt.pie(df['Values'],labels=df['Category'],autopct='%1.1f%%')
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Priyal")
plt.show()
```

```
#4. Boxplot (Outliers Detection)
plt.figure()
sns.boxplot(x=df['Values'])
plt.title("Priyal")
plt.show()
```

```
#5. Heatmap (Correalation)
plt.figure()
corr = df[['Sales','Profit']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Priyal")
print(corr)
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
corr = df[['Values','Sales','Profit']].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True )
plt.title("Priyal")
plt.show()
```

```
# Bubble Plot ( 3 Variable Chart)
import matplotlib.pyplot as plt
plt.scatter(df['Sales'],df['Profit'],s=df['Values']*10) # Size of bubbles
plt.title("Priyal")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales',y='Profit',size='Values' ,hue ='Values', sizes=(50,300),palette='viridis',data=df)
plt.title("Priyal")
plt.show()
```
